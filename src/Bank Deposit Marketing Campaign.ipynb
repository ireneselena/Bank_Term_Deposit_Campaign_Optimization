#### Miss-prediction Conditions and Consequences
**a. Classification:**

**i. Type 1 error (False Positive):**  
Predicted: Customer will make a term deposit  
Actual: Customer does not make a term deposit  

**Consequence:**  
In the case of a false positive, the bank would spend resources on marketing to a customer who ultimately does not make a deposit. For instance, the bank might conduct up to three telemarketing calls, costing approximately €3.0 per call. 

This would amount to a waste of €9.0 per customer on telemarketing alone. Additionally, if email marketing is used, and assuming three emails are sent, the cost per email is roughly €0.011, leading to a further cost of approximately €0.033 per customer. 

There is also the telemarketer hourly salary of €13 which equals to €3.25

Overall, the total wasted cost per false positive could be around €12.283 per customer in addition to the potential loss revenue of €26.75 per customer that will be break down in the Type 2 error (False Negative) calculation below. Thus, total loss for Type 1 error is €39.033

**ii. Type 2 error (False Negative):**  
Predicted: Customer will not make a term deposit  
Actual: Customer does make a term deposit  

**Consequence:**  
In the case of a false negative, the bank would miss out on targeting a potential customer who would have made a deposit. Given that most of the customers are categorized as millennials with an average monthly savings amount of €1,070, the bank could potentially lose out on a significant deposit. 

For example, if a customer intended to make a deposit of €1,070
for a year contract but was not contacted, the bank would lose out on the potential revenue generated from that deposit in which the bank usually use that money to lend it to other customer(s) or businesses with a significantly higher lending rate compared to the deposit interest rate.

Calculated Potential Lost Revenue:

= ($ of deposited amount) * (% of Lending rate - % of Deposit Interest rate)

= €1,070 * (6.0% - 3.5%)

= €1,070 * 2.5%

= €26.75 per customer 


#### Assumptions

Based on this information, the success of our model will be measured by its ability to focus on minimizing false positive while still partially paying attention to its false negatives since the false positive cost is around 1.5x the false negative cost, thereby maximizing the bank’s marketing efficiency and revenue generation.

```python
# Preprocessor for Categorical Columns using One-Hot Encoding
categorical_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),  # Fill missing values with 'unknown'
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  # Apply One-Hot Encoding, ignoring unknown categories
    ])

# Preprocessor for Categorical Columns using Binary Encoding
binary_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),  # Fill missing values with 'unknown'
    ('binary_encoder', BinaryEncoder())  # Apply Binary Encoding to categorical features
    ])

# Preprocessor for the 'month' Column using Ordinal Encoding
month_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing values with the most frequent month
    ('ordinal_encoder', OrdinalEncoder(categories=[['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']]))  # Ordinal encoding based on month order
    ])

# Preprocessor for Other Columns Using Ordinal Encoding
ordinal_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing values with the most frequent value
    ('ordinal_encoder', OrdinalEncoder(categories='auto'))  # Apply Ordinal Encoding with automatic category detection
    ])

# Custom Transformer to Convert Negative Balances to Positive
negbalance_transformer = FunctionTransformer(lambda x: np.abs(x))

# Preprocessor for 'balance' Column
balance_preprocessor = Pipeline(steps=[
    ('negbalance', negbalance_transformer),  # Transform negative balances to positive
    ('imputer', SimpleImputer(strategy='median')),  # Fill missing values with the median
    ('scaler', RobustScaler())  # Scale the balance values using Robust Scaler
    ])

# Preprocessor for Numerical Columns
numeric_preprocessor = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Fill missing values with the median
    ('scaler', RobustScaler())  # Scale numerical values using Robust Scaler
    ])
```

```python
# Feature Creation for Set Feature 3
class Feature_Creation3(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        # Ensure X is a DataFrame
        X = pd.DataFrame(X)

        # 1) Create 'age_group' feature based on age ranges
        X['age_group'] = np.where((X['age'] >= 18)&(X['age']<26), "18-25", 
                   np.where((X['age'] >= 26)&(X['age']<36), "26-35", 
                   np.where((X['age'] >= 36)&(X['age']<46), "36-45", 
                   np.where((X['age'] >= 46)&(X['age']<56), "46-55",
                   np.where((X['age'] >= 56)&(X['age']<66), "56-65", ">65" )))))

        # 2) Convert negative balances to positive
        X['balance'] = np.where(X['balance']<0, X['balance']*-1, X['balance'])
        
        # 3) Create 'balancecategory' feature based on balance ranges
        X['balancecategory'] = np.where((X['balance'] >= 0)&(X['balance']<501), "low",
                               np.where((X['balance'] >= 501)&(X['balance']<2001), "standard",
                               np.where((X['balance'] >= 2001)&(X['balance']<5001), "high", "very high")))

        # 4) Create 'campaigngroup' feature based on campaign counts
        X['campaigngroup'] = np.where((X['campaign'] >= 1)&(X['campaign']<4), "standard",
                             np.where((X['campaign'] >= 4)&(X['campaign']<10), "high", "very high"))
        
        # 5) Create 'lastcontacted' feature based on pdays (days since last contact)
        X['lastcontacted'] = np.where((X['pdays'] >= 0)&(X['pdays']<366), "within 1 year",
                             np.where((X['pdays'] >= 366)&(X['pdays']<731), "within 2 years",
                             np.where(X['pdays'] >= 731, "within 3 years", "never")))

        return X
```

```python
from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, precision_score
import matplotlib.pyplot as plt
import numpy as np

# Setting the X and y
X = Penalize2tunedLGBM.X
y = Penalize2tunedLGBM.y

# Define the precision scorer
precision_scorer = make_scorer(precision_score, average='binary')

# Generate the learning curve
train_sizes, train_scores, test_scores = learning_curve(
    penalize2_tuned_lightgbm, X, y, cv=10, 
    scoring=precision_scorer, train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

# Calculate mean and standard deviation for training and validation scores
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training Precision')
plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Validation Precision')

plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1, color='g')

plt.xlabel('Training Instances')
plt.ylabel('Precision Score')
plt.title('Original Learning Curve (Precision Score)')
plt.legend(loc='best')
plt.show()
```

The original learning curve provides a visual representation of the precision scores for both the training and validation sets as the number of training instances increases.

- Validation Precision: The validation precision starts lower, around 0.68, and steadily improves with additional training data, reaching approximately 0.76. This upward trend indicates that the model's ability to generalize to unseen data improves as it is exposed to more training instances.

- Gap Between Curves: There is a noticeable gap between the training and validation precision curves, which implies that the model is overfitting to some extent. While the training precision remains consistently higher, the increasing validation precision suggests that the model could benefit from more training data or further tuning to reduce this gap and improve generalization.

Overall, the learning curve suggests that while the model performs well on the training set, there is still room for improvement in its ability to generalize to unseen data. Further steps, such as adding more training data, adjusting hyperparameters, or employing regularization techniques, may help reduce the gap between training and validation precision.

Sigmoid calibration also provides improvements but is slightly less effective than isotonic calibration.

#### **How Does this Final Model Works**

The final model utilized is a Light Gradient Boosting Machine (LGBM), which is a powerful and efficient algorithm for classification tasks. Here's a breakdown of how this model works:

1. Light Gradient Boosting Machine (LGBM) Overview:

- Gradient Boosting: LGBM is an implementation of gradient boosting, a technique that builds an ensemble of decision trees in a sequential manner. Each new tree attempts to correct the errors made by the previous trees.
- Efficiency: LGBM is optimized for speed and performance. It uses techniques like histogram-based decision trees and leaf-wise growth, which allow it to handle large datasets with high-dimensional features efficiently.

2. Calibration Using Sigmoid:

- Probability Calibration: The raw probabilities predicted by the LGBM model are often not perfectly calibrated, meaning they may be overconfident or underconfident. To address this, the model's output probabilities are calibrated using the sigmoid method.
- Sigmoid Calibration: This technique applies a logistic function to the predicted probabilities, adjusting them so that they better reflect the true likelihood of the positive class. Sigmoid calibration is particularly useful when the model's raw predictions are not well-aligned with the actual class probabilities.
- Improved Reliability: By calibrating the probabilities, the final model provides more reliable probability estimates, which are crucial for applications where the decision-making process depends on accurate probability thresholds.

3. Model Predictions:

- Training and Validation: The model was trained and validated using historical data, where it learned to identify patterns and relationships between the features and the target variable (the outcome to be predicted).
- Predictions: Once trained, the model can be used to predict the likelihood of the positive class (e.g., whether a customer will take a specific action) for new, unseen data. These predictions are expressed as probabilities, allowing for a nuanced understanding of the model’s confidence in each prediction.
- Calibration Impact: The calibrated probabilities provide a more accurate representation of the model's confidence in its predictions, leading to better decision-making based on these probabilities.

4. Performance Metrics:

- Precision and Recall: The model's performance is evaluated using precision, recall, and other metrics, ensuring that it not only makes accurate predictions but also balances the trade-off between identifying positive instances and avoiding false positives.
- Calibration Effectiveness: The effectiveness of the sigmoid calibration is evident in the improved alignment between predicted probabilities and observed outcomes, which enhances the model's reliability in real-world applications.

Summary:
The final model, based on LGBM with sigmoid calibration, is a robust and efficient tool for making well-calibrated predictions. It leverages the power of gradient boosting to build a strong classifier while ensuring that the predicted probabilities are reliable and interpretable. This combination makes the model suitable for deployment in scenarios where accurate probability estimates are crucial for decision-making.

#### **Final Model Limitation**

While the final model, based on the Light Gradient Boosting Machine (LGBM) with sigmoid calibration, performs well on the data it was trained and validated on, there are certain limitations that need to be acknowledged:

- Dependence on Historical Data:
    The model was trained using historical data, meaning it learned patterns and relationships specific to the data available during the training period. This can limit the model's ability to generalize to future data, particularly if the patterns in the data change over time or if new, unseen trends emerge.

- Numerical Feature Ranges:
    The model was trained on numerical columns with specific ranges of values:
        - Age: 18 to 95 years
        - Balance: -6847 to 66653 units
        - Campaign: 1 to 63 contacts
        -   Pdays: -1 to 854 days

    If, in the future, the data contains numerical values outside these ranges, the model's predictions may become unreliable. This is because the model has not seen such values during training and may not correctly extrapolate its predictions.

- Categorical Feature Values:
    The categorical columns in the training data have a fixed set of possible values. The model learned to make predictions based on these specific categories.

    If new, previously unseen categorical values are introduced in future data, the model may struggle to handle them effectively, leading to potentially inaccurate or unreliable predictions.

- Model Generalization:
    The model's generalization ability is inherently limited by the diversity and representativeness of the training data. Any significant deviations from the conditions present in the training data could affect the model's performance, making it less reliable in those cases.

In summary, while the final model is robust within the context of the training data, it is essential to be cautious when applying it to future data that may fall outside the ranges or categories it was trained on. Regular updates and retraining of the model, along with careful monitoring of its performance, will be necessary to maintain its reliability and effectiveness over time.
